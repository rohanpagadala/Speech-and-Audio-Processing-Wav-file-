```python
import numpy as np
import matplotlib.pyplot as plt
import librosa
import scipy.io.wavfile as wav

fs, wav_data = wav.read('/content/name.wav')

fwav = wav_data / 32000.0

spec = librosa.stft(fwav)

plt.figure(figsize=(10, 6))
plt.imshow(np.abs(spec), aspect='auto', origin='lower', cmap='inferno')
plt.colorbar(label="Magnitude (dB)")
plt.title("Spectrogram of Audio")
plt.ylabel("Frequency (Hz)")
plt.xlabel("Time (s)")
plt.show()
```
